```python
from sklearn.model_selection import train_test_split
```

```python
df3.fillna(0,inplace=True)
X=df3.drop(['Mental disorders'],axis=1)
y=df3['Mental disorders']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
predict1=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse_lr=mean_squared_error(y_test, predict1)
rmse_lr=(np.sqrt(mean_squared_error(y_test, predict1)))
r2_lr=r2_score(y_test, predict1)
print(mse_lr)
print(rmse_lr)
print(r2_lr)
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
```

```python
rfr.fit(X_train,y_train)
```

```python
predict2=rfr.predict(X_test)
```

```python
mse_rfr=mean_squared_error(y_test, predict2)
rmse_rfr=(np.sqrt(mean_squared_error(y_test, predict2)))
r2_rfr=r2_score(y_test, predict2)
print(mse_rfr)
print(rmse_rfr)
print(r2_rfr)
```

# DecisionTree Regresssor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
predict3=dtr.predict(X_test)
```

```python
mse_dtr=mean_squared_error(y_test, predict3)
rmse_dtr=(np.sqrt(mean_squared_error(y_test, predict3)))
r2_dtr=r2_score(y_test, predict3)
print(mse_dtr)
print(rmse_dtr)
print(r2_dtr)
```

# SVR


```python
from sklearn.svm import SVR
```

```python
svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
predict4=svr.predict(X_test)
```

```python
mse_svr=mean_squared_error(y_test, predict4)
rmse_svr=(np.sqrt(mean_squared_error(y_test, predict4)))
r2_svr=r2_score(y_test, predict4)
print(mse_svr)
print(rmse_svr)
print(r2_svr)
```
